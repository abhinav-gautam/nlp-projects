# Translator - to Spanish


```python
from textblob import TextBlob
from textblob import Word
```

```python
# Reading file
file = open('article.txt','r')
file_content = TextBlob(file.read())
```

```python
# Sentences in paragraph
for sentence in file_content.sentences:
    print(sentence+"\n")
```

```python
# Words in paragraph
for word in file_content.words:
    print(word)
```

```python
# Nouns in paragraph
for word,tag in file_content.pos_tags:
    if 'NN' in tag:
        print(word)
```

```python
# Spelling Correction
file_corrected = file_content.correct()
file_corrected
```

```python
# Translate into spanish
file_corrected.translate(to="es")
```
